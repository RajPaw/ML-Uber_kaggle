```python
##In the Below line of code we try to find the Starting destination with highest number of rides.
```

```python
starting_point = uber_df["START*"].dropna()
df = pd.DataFrame(starting_point.value_counts())
df.sort_values(["START*"], ascending = False)
df = df.reset_index()
df = df.rename(columns = {'index':'starting_destination', 'START*':'Count'})
df.loc[df['Count'] == max(df['Count'])]
```

```python
starting_point = uber_df["START*"].dropna()
df = pd.DataFrame(starting_point.value_counts())
df.sort_values(["START*"], ascending=True, inplace=True)
df.reset_index(inplace=True)
df.rename(columns={'index':'starting_destination', 'START*':'Count'}, inplace=True)
result = df[df['Count'] == min(df['Count'])]

print(result)
```

```python
starting_point = uber_df["START*"].dropna()
df = pd.DataFrame(starting_point.value_counts())

df.reset_index(inplace=True)
df.rename(columns={'index':'starting_destination', 'START*':'Count'}, inplace=True)

df
```

```python
stoping_point = uber_df["STOP*"].dropna()
df = pd.DataFrame(stoping_point.value_counts())

df.reset_index(inplace=True)
df.rename(columns={'index':'stoping_destination', 'STOP*':'Count'}, inplace=True)

df
```

```python
## In the below Code I found out the most frequent route taken by Uber rides.
```

```python
df = uber_df.dropna()
df = pd.DataFrame(df.groupby(['START*', 'STOP*']).size())
df = df.rename(columns = {0:'Count'})
df = df.sort_values(['Count'], ascending = False)
df.loc[df['Count'] == max(df['Count'])]
```

```python
## Here, we printed the all type of purpose of the trip
```

```python
print(np.array(uber_df['PURPOSE*'].dropna().unique()))
uber_df['MILES*'].groupby(uber_df['PURPOSE*']).sum()
```

```python
max_miles_purpose = uber_df.groupby('PURPOSE*')['MILES*'].sum().idxmax()
max_miles_count = uber_df.groupby('PURPOSE*')['MILES*'].sum().max()
print(f"The category with the maximum miles covered is: {max_miles_purpose}")
print(f"The total miles covered in that category is: {max_miles_count}")
```

```python
df = pd.DataFrame(uber_df['MILES*'].groupby(uber_df['PURPOSE*']).sum())
df.plot(kind = 'bar')
plt.show()
```

```python
import matplotlib.pyplot as plt

df = pd.DataFrame(uber_df['MILES*'].groupby(uber_df['PURPOSE*']).sum())

plt.figure(figsize=(8, 8))
plt.pie(df['MILES*'], labels=df.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Miles by Purpose')
plt.axis('equal')

plt.show()
```

```python

df = pd.DataFrame(uber_df['CATEGORY*'].value_counts())
plt.figure(figsize=(8, 8))
plt.pie(df['CATEGORY*'], labels=df.index, autopct='%1.1f%%', startangle=140)
plt.title('CATEGORICAL TRIP')
plt.axis('equal')

plt.show()
```
